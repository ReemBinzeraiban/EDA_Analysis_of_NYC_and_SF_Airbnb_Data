#   <a name='imports' >Data Cleaning</a>

```python
mldf = sf_df[['city','neighbourhood', 'neighbourhood_cleansed', 'room_type', 'price','latitude','longitude','minimum_nights','availability_365','number_of_reviews']]
```

```python
mldf.head()
```

```python
# # Drop null values; check what datatype each column contains:

mldf.dropna(inplace = True)

mldf.info()
```

```python
mldf.isnull().sum()
```

```python
#This looks good. We dont need to drop any duplicated data.

sum(mldf.duplicated())
```

```python
#check duplicates
mldf = mldf.drop_duplicates()
print(mldf)
```

```python
sum(mldf.duplicated())
```

```python
mldf['city'].value_counts()
```

```python
#We want only listings located in San Francisco
mldf.loc[mldf['city'].str.contains('San Francisco ')]
```

```python
# # Correct the 'San Francisco' name variants to read only 'San Francisco':

mldf.loc[mldf['city'] == 'Noe Valley - San Francisco', 'city'] = 'San Francisco'
mldf.loc[mldf['city'] == 'San Francisco, Hayes Valley', 'city'] = 'San Francisco'
mldf.loc[mldf['city'] == '', 'city'] = 'San Francisco'
mldf.loc[mldf['city'] == 'San Francisco ', 'city'] = 'San Francisco'
mldf.loc[mldf['city'] == 'San Jose', 'city'] = 'San Francisco'
mldf.loc[mldf['city'] == 'Brisbane', 'city'] = 'San Francisco'
mldf.loc[mldf['city'] == 'Daly City', 'city'] = 'San Francisco'
```

```python
# Confirm the corrections using value_counts method:

mldf['city'].value_counts()
```

```python
# # Convert the values in price column to floats:
mldf['price'] = mldf['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
mldf.describe()
```

```python
#The min price value is zero. This make no sense. Let's check for prices < 1
select_price = mldf.loc[mldf['price'] <1]
select_price
```

# Whta is the room type and price for each states?

```python
# whta is the room type and price for each states?

fig=plt.figure(figsize=(10,33))
fig.subplots_adjust(hspace=0.15, wspace=0.75)

plt.subplot(3,2,1)
df[df["room_type"] == "Entire home/apt"].groupby(["neighbourhood_group"])['price'].median().sort_values(ascending=True).plot.barh(color="skyblue")
plt.xticks(rotation=50, size=10, ha='right')
plt.xlabel("Median Price in US$")
plt.title(label='Entire home/apt Median Prices in NYC')

plt.subplot(3,2,2)
mldf[mldf["room_type"] == "Entire home/apt"].groupby(["neighbourhood_cleansed"])['price'].median().sort_values(ascending=True).plot.barh()
plt.xticks(rotation=50, size=10, ha='right')
plt.xlabel("Median Price in US$")
plt.ylabel("")
plt.title(label='Entire home/apt Median Prices in SF')

plt.subplot(3,2,3)
df[df["room_type"] == "Private room"].groupby(["neighbourhood_group"])['price'].median().sort_values(ascending=True).plot.barh(color="skyblue")
plt.xticks(rotation=50, size=10, ha='right')
plt.xlabel("Median Price in US$")
plt.title(label='Private room Median Prices in NYC')

plt.subplot(3,2,4)
mldf[mldf["room_type"] == "Private room"].groupby(["neighbourhood_cleansed"])['price'].median().sort_values(ascending=True).plot.barh()
plt.xticks(rotation=50, size=10, ha='right')
plt.xlabel("Median Price in US$")
plt.ylabel("")
plt.title(label='Private room Median Prices in SF')

plt.subplot(3,2,5)
df[df["room_type"] == "Shared room"].groupby(["neighbourhood_group"])['price'].median().sort_values(ascending=True).plot.barh(color="skyblue")
plt.xticks(rotation=50, size=10, ha='right')
plt.xlabel("Median Price in US$")
plt.title(label='Shared room Median Prices in NYC')

plt.subplot(3,2,6)
mldf[mldf["room_type"] == "Shared room"].groupby(["neighbourhood_cleansed"])['price'].median().sort_values(ascending=True).plot.barh()
plt.xticks(rotation=50, size=10, ha='right')
plt.xlabel("Median Price in US$")
plt.ylabel("")
plt.title(label='Shared room Median Prices in SF')
plt.savefig("RT.png")
plt.show();
```
